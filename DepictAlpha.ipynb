```python
import numpy as np
from pathlib2 import Path

import visualization
import random
import matplotlib.pyplot as plt
import os


NUM_SECTOR = 8
NUM_BIN = 2
BATCH_SIZE = 4
NUM_WORKERS = 2

dataset = "./dataset_small/"
root_path  = dataset+"training/"
data_label = root_path + 'label_2/'
data_img   = root_path + 'image_2/'
calib_path = root_path + "calib/"
```

```python
instance = "000005"
img_path = data_img +instance+".png"
label_path = data_label +instance+".txt"
img = plt.imread(img_path)
calib_pth = os.path.join(calib_path, instance + ".txt")

import math
for line in open(label_path):
    line_split = line.split(' ')
    alpha = float(line_split[3])
    print('alpha: ', alpha)
    x = float(line_split[11])
    y = float(line_split[12])
    z = float(line_split[13])
    theta = alpha + math.atan2(x, z) # the formula from gs 3d
    print('rot_y_alpha: ', theta)
    print('rot_y: ', line_split[14])
    img = visualization.make_alpha_arrow(calib_pth, line, img, alpha_pred=alpha)
plt.figure(figsize=(150, 16))
plt.imshow(img)
```
