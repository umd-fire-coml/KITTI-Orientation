# **Monocular 3D-Object Detection**

---



The goal of our project is to accurately predict the orientation (blue arrows above) of vehicles only using a single camera as input. Previous works [Multi-bin] have taken complicated approaches, utilizing 2 overlapping bins that are slightly greater than 180 degrees to allow for regression using this overlap and the overall bin confidence. Most modern works have opted for simpler approaches, directly regressing orientation using the input image's features [GS3D, Joint Monocular ..., SMOKE]. We propose a promising method of determining orientation, Tricosine .... 

We predict three cosine values that determine the main orientation of the object and the direction/magnitude of the offset from the main direction

---

```python
import tensorflow.keras as keras
import tensorflow as tf
from backbone_xception import Xception_model
from add_output_layers import add_output_layers
from tensorflow.keras import Input, Model
from loss_function import *
import numpy as np
import data_processing_val as dp
import os, argparse, time, pickle
from pathlib2 import Path

import visualization
import random
import matplotlib.pyplot as plt

NUM_SECTOR = 8
NUM_BIN = 2
BATCH_SIZE = 4
NUM_WORKERS = 2

root_path  = "./dataset/training/"
data_label = root_path + 'label_2/'
data_img   = root_path + 'image_2/'
calib_path = root_path + "calib/"

orientation = 'tricosine'
num_epoch = 82


if orientation == 'alpha_sector' or orientation == 'rot_y_sector':
    sect_add = "_{}".format(NUM_SECTOR)
else:
    sect_add = ""
    
# sect_add = "_{}".format(NUM_SECTOR) if orientation == 'alpha_sector' or orientation == 'rot_y_sector' else ""
weights_path = "./weights/{o}{sect}/{o}_model.{e}.h5".format(o=orientation,
                                                             e=num_epoch,
                                                             sect=sect_add)
```

### Initialize KITTI Data Loader

---

```python
#Build data loader for KITTI data
generator = dp.KittiGenerator(label_dir= data_label, 
                              image_dir= data_img,
                              mode = 'val',
                              sectors = NUM_SECTOR, 
                              batch_size = BATCH_SIZE,
                              orientation_type = orientation)

# Get one batch
gen_output = generator[0]
```

## Build Model

---

```python
inputs = Input(shape=(224, 224, 3))
stage1 = Xception_model(inputs, pooling='avg')
stage2 = add_output_layers(orientation, stage1, NUM_SECTOR,NUM_BIN)
model  = Model(inputs=inputs, outputs=stage2)

model.compile(loss=loss_func(orientation), optimizer='adam')
model.load_weights(weights_path)

pred = model.predict(x=gen_output[:2],
                     verbose=1,
                     workers=NUM_WORKERS)
```



### Visualize Model
